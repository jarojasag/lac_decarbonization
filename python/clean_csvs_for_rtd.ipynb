```python
all_csvs = [x for x in os.listdir(sa.dir_attribute_tables) if (".csv" in x) and ("table_varreqs" in x)]
dict_tables = {}
field_drop = ""

dict_fields = {
    "Internal Model Variable": (0, int),
    "Default LHS Scalar Minimum at Final Time Period": (1, float),
    "Default LHS Scalar Maximum at Final Time Period": (1, float),
    "Dashboard": ("", str),
    "Reference": ("", str),
    "Data Link": ("", str)
}
def_val = 0

ordering = "append"
force_default_override_if_exists = False

ord_by_cat = [
    "Variable Type",
    "Variable",
    "Information",
    "Variable Schema",
    "Dashboard",
    "Notes",
    "Data Link",
    "Reference",
    "Internal Model Variable",
    "Default LHS Scalar Minimum at Final Time Period",
    "Default LHS Scalar Maximum at Final Time Period"
]

ord_by_partial_cat = [
    "Variable Type",
    "Variable",
    "Information",
    "Variable Schema",
    "Dashboard",
    "Categories",
    "Notes",
    "Data Link",
    "Reference",
    "Internal Model Variable",
    "Default LHS Scalar Minimum at Final Time Period",
    "Default LHS Scalar Maximum at Final Time Period"
]

for csv in all_csvs:
    print(csv)
    fl = pd.read_csv(os.path.join(sa.dir_attribute_tables, csv), skipinitialspace = True)
    flds = list(fl.columns)
    
    for new_field in dict_fields.keys():
                                
        def_val = dict_fields[new_field][0]
        type_new = dict_fields[new_field][1]
        
        if (new_field not in fl.columns):
            fl[new_field] = def_val
            print(f"column {new_field} not found in {os.path.basename(csv)}")
        elif force_default_override_if_exists:
            fl[new_field] = def_val
        fl[new_field].fillna(def_val)
        
        fl[new_field] = np.array(fl[new_field]).astype(type_new)
            
    if "_partial_category_" in csv:
        fl = fl[ord_by_partial_cat]
    else:
        fl = fl[ord_by_cat]
    fl = fl.sort_values(by = ["Variable Type", "Variable"]).reset_index(drop = True)

    dict_tables.update({csv: fl})
    print("done")
    
```
